# ***ricerca di stringhe***

```python
#ricerca di stringhe
movies[(movies["Film"]=="Porno")]
```

```python
movies[movies["Film"].str.contains("Porno")]  #permette di cercare tutto qullo che contine(lettere , numeri) 
# 'contains' questo permette tuttio ciÃ²
```

```python
movies[movies["Film"].str.contains("the")] #permette di cercare tutto qullo che contine(lettere , numeri) 
# 'contains' questo permette tuttio ciÃ²
```

```python
#cerca la prima lettere in una stringa
movies[movies["Film"].str.startswith("T")]  #'startswith' questo permette tuttio ciÃ² 
```

```python
movies[movies["Film"].str.endswith("on")]  #visulizza la stringa in cui contiene le ultime lettere 
```

```python
#visualizare i film in ordine alfabetico
movies.sort_values(by=["Film"])
```
